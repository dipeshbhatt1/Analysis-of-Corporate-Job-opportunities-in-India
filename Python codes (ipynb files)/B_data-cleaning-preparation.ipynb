```python
import pandas as pd
import re
```

```python
df = pd.read_csv(r"C:\Users\HP\OneDrive\Desktop\MASAI\PROJECT-20230705T045048Z-001\PROJECT\Web Scraping\Misc\Job_details500.csv", index_col = 0)
```

```python
df.head()
```

```python
df.shape
```

```python
df['designation'].nunique()
```

```python
job_id = []

for i in range(1, len(df)+1):
    job_id.append(f'J{i}')
```

```python
len(job_id)
```

```python
job_id = pd.DataFrame(data=job_id, columns=['job_id'])
```

```python
df = pd.concat([df, job_id], axis=1)
```

```python
df.head(2)
```

```python
df[df.duplicated()]
```

```python
df['designation'].nunique()
```

```python
df['designation'] = df['designation'].apply(lambda x: " ".join(i for i in x.split() if i not in ['(Drive)']))
```

```python
df['location'].unique()
```

```python
df = df[(df['location'] != 'United Arab Emirates (UAE)') & (df['location'] != 'Sri Lanka')]
```

```python
df.shape
```

```python
df['estab_year'].unique()
```

```python
df = df[df['estab_year'] != '0 - 10 employees']
```

```python
df['estab_year'] = df['estab_year'].apply(lambda x: x.split()[2])
```

```python
df['employees_count'].unique()
```

```python
df['employees_count'] = df['employees_count'].apply(lambda x: (re.sub("[^0-9-\s]+", "", x)).strip())
```

```python
df['employees_count'].unique()
```

```python
df['employees_count'].replace('1000', '1000+', inplace=True)
```

```python
df['experience_req'].sort_values().unique()
```

```python
df['experience_req'].replace(['0-0 Years', '0-1 Years', '0-2 Years', '0-3 Years', '0-4 Years'], 'Fresher', inplace=True)
```

```python
df['job_description'] = df['job_description'].apply(lambda x: x.split("→")[0].split(":")[1].strip())
```

```python
df['job_description'].unique()
```

```python
df['skills'] = df['skills'].str.replace("\n", ", ")
```

```python
df['skills'].unique()
```

```python
df['location'].unique()
```

```python
df.head()
```

```python
df.drop('hr_name', axis=1, inplace=True)
```

```python
for i in df.columns:
    df[i] = pd.to_numeric(df[i], errors = 'ignore')
```

```python
df.dtypes
```

```python
df.rename(columns={'job_description':'industry'}, inplace=True)
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df = df[~df['skills'].isnull()]
```

```python
df.shape
```

```python
for i in df.columns:
    if df[i].dtypes != "O":
        df[i] = df[i].fillna(df[i].median())
```

```python
df.isnull().sum()
```
